```python
# 替所有欄位取名的階段
column_names = ["orignal_time","position","username"]
mos_list = ['Status','Graphic','Interactive','SelfAssessment','Opinion']

wks = wks_list[0]
df = wks.get_as_df()

for i in range(10):
    for mos in mos_list:
        column_names.append(f'round{i}_{mos}')
# print(column_names)

# 欄位名稱命名
df.columns = column_names
# 欄位拆分
df[['date','midday','time']] = df.orignal_time.str.split(' ',expand=True)
# df.info()
df.head()
```

```python
# 取出所有座位的資料並合併
temp_df = pd.DataFrame()
df['order'] = ''
for i in range(1,5):
    tdf = df.query(f'position == {i}')
    j = 1
    prev_date = ''
    for index, row in tdf.iterrows():
        date = ''.join(row['date'].split('/'))
        if row['date'][-2] == '/':
            date = date[:-1] + '0' + date[-1]
        if date != prev_date:
            j = 1
            prev_date = date
        else:
            j += 1
        filename = os.path.join('qoestorage',f'vm{i}','allmanualConfig', f'{date}_{j}_{i}.txt')
        if os.path.exists(filename):
            print("檔案存在", filename)
            with open(filename, 'r') as f:
                lines = []
                for line in f:
                    lines.append(line.strip())
                order = ','.join(lines)
                print(order)
            tdf.loc[index, 'order'] = order
        else:
            print(f'file {filename} not exists')
        # print(index,date,j)
    print()
    # tdf['time'] = pd.to_datetime(tdf['time'])
    temp_df = pd.concat([temp_df,tdf], ignore_index=True)
df = temp_df
```

```python
# 排序過的 DataFrame 從這裡開始
ordered_df = df[['time','order']]
for i in range(10):
    for mos in mos_list:
        ordered_df[f'round{i}_{mos}'] =''
ordered_df
```

```python
empty_control_variable = {
    'latency': 0,
    'packet_loss': 0,
    'bandwidth': 0
}
latency_list = [30,100,200]
packet_loss_list = [15,20,25]
bandwidth_list = [1920,2560,32000] # 15,20,250 M
control_variable_list = empty_control_variable.keys()
expand_df = pd.DataFrame(columns=['time','position','username','round']+[mos for mos in mos_list]+[v for v in control_variable_list])

for index, row in df.iterrows():
    orderstr = row['order']
    # print(type(order))
    if orderstr != '':
        order_list = orderstr.split(',')
        rowdata = {
            'time': df['orignal_time'][index],
            'position': df['position'][index],
            'username': df['username'][index],
            'round': 0
        }
        rowdata.update(empty_control_variable)
        for v in control_variable_list:
            rowdata[v] = 0
        for mos in mos_list:
            rowdata[mos] = df[f"round{rowdata['round']}_{mos}"][index]
        expand_df = expand_df.append(rowdata,ignore_index=True)
        for order in order_list:
            rowdata['round'] += 1
            rowdata.update(empty_control_variable)
            if order == '1':
                rowdata['latency'] = latency_list[0]
            elif order == '2':
                rowdata['latency'] = latency_list[1]
            elif order == '3':
                rowdata['latency'] = latency_list[2]
            elif order == '4':
                rowdata['packet_loss'] = packet_loss_list[0]
            elif order == '5':
                rowdata['packet_loss'] = packet_loss_list[1]
            elif order == '6':
                rowdata['packet_loss'] = packet_loss_list[2]
            elif order == '7':
                rowdata['bandwidth'] = bandwidth_list[0]
            elif order == '8':
                rowdata['bandwidth'] = bandwidth_list[1]
            elif order == '9':
                rowdata['bandwidth'] = bandwidth_list[2]
            for mos in mos_list:
                rowdata[mos] = df[f"round{rowdata['round']}_{mos}"][index]
            expand_df = expand_df.append(rowdata,ignore_index=True)
from google.colab import files
expand_df.to_csv('expand_mos.csv', encoding = 'utf-8-sig') 
# 下載檔案
# files.download('expand_mos.csv')
```

```python
# 亂數位置調整回原樣 (只能執行一次)
for index, row in ordered_df.iterrows():
    order = row['order']
    # print(type(order))
    if order != '':
        order_list = order.split(',')
        for mos in mos_list:
            # 取出最佳狀況
            ordered_df.loc[index, f'round0_{mos}'] = df.loc[index, f'round0_{mos}']
            i = 1
            for o in order_list:
                # 將亂數調整回相同順序
                if o != '':
                    ordered_df.loc[index, f'round{o}_{mos}'] = df.loc[index, f'round{i}_{mos}']
                    i += 1
        # print(order_list)
ordered_df
```
