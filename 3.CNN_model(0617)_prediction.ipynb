```python
# test_filenames = os.listdir("/content/test_vm3")
b = os.listdir(r"/content/test_vm1")
b.sort(key = lambda x:x.split('.')[0][16:50].replace('_','').replace('(','').replace(')','').replace('-',''))

test_df = pd.DataFrame({
    'file': b
})
nb_samples = test_df.shape[0]
```

```python
test_df.head()
```

```python
FAST_RUN = False
IMAGE_WIDTH=128
IMAGE_HEIGHT=128
IMAGE_SIZE=(IMAGE_WIDTH, IMAGE_HEIGHT)
IMAGE_CHANNELS=3
batch_size=15
```

```python
test_gen = ImageDataGenerator(rescale=1./255)
test_generator = test_gen.flow_from_dataframe(
    test_df,
    "/content/test_vm1/",
    x_col='file',
    y_col=None,
    class_mode=None,
    target_size=IMAGE_SIZE,
    batch_size=batch_size,
    shuffle=False
)
```

```python
predict = model.predict(test_generator, steps=np.ceil(nb_samples/batch_size),
                                 verbose=1)
```

```python
test_df['category'] = np.argmax(predict, axis=-1)
```

```python
#0=gaming,1=nogaming
test_df['category'].value_counts().plot.bar()
```

```python
sample_test = test_df.head(18)
sample_test.head()
plt.figure(figsize=(24, 24))
for index, row in sample_test.iterrows():
    filename = row['file']
    category = row['category']
    img = load_img("/content/test_vm1/"+filename, target_size=IMAGE_SIZE)
    plt.subplot(6, 3, index+1)
    plt.imshow(img)
    plt.xlabel(filename + '(' + "{}".format(category) + ')' )
plt.tight_layout()
plt.show()
```
